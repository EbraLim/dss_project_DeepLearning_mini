```python
# roi_extract
o_list_3rd = ox.roi_extract(o_list_3rd)
x_list_3rd = ox.roi_extract(x_list_3rd)
```

```python
# train_test_split
X_train, X_test, y_train, y_test = ox.concat_and_split(o_list_3rd, x_list_3rd)
```

```python
ann_3rd = ox.create_ANN_model()
cnn_3rd = ox.create_CNN_model()
```

```python
ann_3rd_hist = ox.get_result_ann(X_train, X_test, y_train, y_test, ann_3rd, batch_size=10)
```

```python
cnn_3rd_hist = ox.get_result_cnn(X_train, X_test, y_train, y_test, cnn_3rd, batch_size=10)
```

```python
# ANN에서 틀린 데이터 살펴보기
predicted_result = ann_3rd.predict(X_test)
predictd_labels = np.argmax(predicted_result, axis=1)

wrong_result = []
for n in range(0, len(y_test)):
    if predictd_labels[n] != y_test[n]:
        wrong_result.append(n)

len(wrong_result)        
```

```python
# ANN에서 틀린 데이터 살펴보기
plt.figure(figsize=(12,3))

for idx, n in enumerate(wrong_result):
    plt.subplot(1, 2, idx+1)
    plt.imshow(X_test[n], cmap="gray", interpolation='nearest')
    plt.title("Label: " + str(y_test[n]) + ' / Predict: ' + str(predictd_labels[n]))
    plt.grid(False)
#     plt.axis('off')
    
plt.tight_layout()
plt.show()
```

```python
# 다른 것들도 많이 뭉개졌는지 확인 -> 대부분 그렇지 않음
# 붓펜으로 그린 그림 일부에서 나타나는 것으로 보임
plt.figure(figsize=(16, 8))
tmp = np.random.randint(0,64,8)
for idx, n in enumerate(tmp):
    plt.subplot(2, 4, idx+1)
    plt.imshow(X_test[n], cmap='gray', interpolation='nearest')
    plt.title("Label: " + str(y_test[n]) + ' / Predict: ' + str(predictd_labels[n]))
    plt.grid(False)
    
plt.tight_layout()
plt.show()
```

```python
# CNN에서 틀린 데이터 살펴보기
X_test_reshaped = X_test.reshape(X_test.shape + (1,))
predicted_result_cnn = cnn_3rd.predict(X_test_reshaped)
predictd_labels_cnn = np.argmax(predicted_result_cnn, axis=1)

wrong_result_cnn = []
for n in range(0, len(y_test)):
    if predictd_labels_cnn[n] != y_test[n]:
        wrong_result_cnn.append(n)

len(wrong_result_cnn)
```

```python
# CNN에서 틀린 데이터 살펴보기
plt.figure(figsize=(6,6))

for idx, n in enumerate(wrong_result_cnn):
    plt.subplot(1, 2, idx+1)
    plt.imshow(X_test_reshaped[n], cmap="gray", interpolation='nearest')
    plt.title("Label: " + str(y_test[n]) + ' / Predict: ' + str(predictd_labels_cnn[n]))
    plt.grid(False)
    plt.axis('off')
    
plt.tight_layout()
plt.show()
```
